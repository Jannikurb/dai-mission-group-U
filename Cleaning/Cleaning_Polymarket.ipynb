Interpretation of Prices on Polymarket: A Polymarket share pays $1 if the event occurs and $0 otherwise. Therefore, prices are bounded between 0 and 1 and can be interpreted as market-implied probabilities. 

So prices lower than 0 and higher than 1 are wrong and need to be removed.

```python
# Set invalid bid/ask values (<0 or >1) to NaN

df_markets_small.loc[
    (df_markets_small["bestBid"] < 0) | (df_markets_small["bestBid"] > 1),
    "bestBid"
] = pd.NA

df_markets_small.loc[
    (df_markets_small["bestAsk"] < 0) | (df_markets_small["bestAsk"] > 1),
    "bestAsk"
] = pd.NA
```

why is the count of spread 200.000 if the count of bestBid is only 92.809?

```python
# check markets where bestBid is NaN

df_markets_small[
    df_markets_small["bestBid"].isna()
][["question", "bestBid", "bestAsk", "spread"]].head(10)
```

So this spread variable interpret NAN in bestBid as 0. This is wrong because when bestBid is NAN, then there is no bid at all. Therefore I´m creating a new variable spread where NAN is interpreted as NAN and not 0.

## 4.1 spread

```python
# before creating a new variable spread, we rename the existing spread variable to "spread_poly" to avoid confusion

df_markets_small.rename(columns={"spread": "spread_poly"}, inplace=True)

# check
df_markets_small.columns
```

```python
# create a new variable spread which is the difference between bestAsk and bestBid

df_markets_small["spread"] = (
    df_markets_small["bestAsk"] - df_markets_small["bestBid"]
)

# test if there are as many non-null entrys as in bestBid
df_markets_small["spread"].describe()
```

## 4.4 Market Uncertainty

The uncertainty measure is calculated as:

Uncertainty = 1 - |p_1 - p_2|

where \(p_1\) and \(p_2\) are the probabilities of the two outcomes.

The measure ranges from 0 to 1:

- Uncertainty = 1: Maximum uncertainty (50% vs. 50%)
- Uncertainty = 0: No uncertainty (100% vs. 0%)

Therefore, higher values indicate that the market is more uncertain about the outcome, while lower values indicate a stronger market consensus.

Markets with more than two outcomes (40 observations, 0.02% of the sample) were assigned missing values (`NaN`) because the uncertainty measure is designed for binary markets only.

```python
# create a new variable uncertainty

import ast
import numpy as np

def calculate_uncertainty(x):
    probs = [float(i) for i in ast.literal_eval(x)]

    if len(probs) != 2:
        return np.nan

    return 1 - abs(probs[0] - probs[1])

df_markets_small["uncertainty"] = (
    df_markets_small["outcomePrices"]
    .apply(calculate_uncertainty)
)

# Test
df_markets_small[["outcomePrices", "uncertainty"]].head()
```
